```python
import seaborn as sns
import matplotlib.pyplot as plt

# PDF, Histograms, Boxplots, Countplots
# Example for numerical column 'Salary'
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True)
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

# - The salary distribution is right-skewed with most salaries concentrated towards the lower end.


```python
# Example for categorical column 'Gender'
plt.figure(figsize=(8, 5))
sns.countplot(df['Gender'])
plt.title('Countplot of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# - There are more male individuals in the dfset compared to female individuals.

```python
# Outliers
# Example for numerical column 'Salary'
plt.figure(figsize=(8, 5))
sns.boxplot(df['Salary'])
plt.title('Boxplot of Salary')
plt.xlabel('Salary')
plt.show()
```

# - The boxplot suggests the presence of outliers in the Salary column.

```python
#Bivariate Analysis
```

```python
# Discover the relationships between numerical columns
# Example: Scatter plot between 'Salary' and '10percentage'
plt.figure(figsize=(8, 5))
sns.scatterplot(x='10percentage', y='Salary', df=df)
plt.title('Scatter plot of 10percentage vs Salary')
plt.xlabel('10percentage')
plt.ylabel('Salary')
plt.show()
```

# - There seems to be a positive correlation between 'Salary' and '10percentage'.

```python
# Identify the patterns between categorical and numerical columns
# Example: Boxplot of 'Salary' vs 'Gender'
plt.figure(figsize=(8, 5))
sns.boxplot(x='Gender', y='Salary', df=df)
plt.title('Boxplot of Salary vs Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()
```

# - The boxplot suggests a difference in salary between genders, with males generally earning higher.


```python
# Identify relationships between categorical and categorical columns
# Example: Stacked bar plot of 'Gender' vs 'Specialization'
gender_spec = pd.crosstab(df['Gender'], df['Specialization'])
gender_spec.plot(kind='bar', stacked=True)
plt.title('Stacked bar plot of Gender vs Specialization')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# - The stacked bar plot shows the distribution of specializations among genders.
